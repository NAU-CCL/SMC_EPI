```python

'''Moving average of the data, using the discrete convolution in numpy, investigated to determine if convergence is improved'''


import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
np.set_printoptions(suppress=True)

state = 'Arizona'

df = pd.read_csv('../datasets/state_case_JHU.csv',index_col=False)
state_data = df.loc[df['state'] == state]
state_data = state_data.reset_index(drop=True)
state_data = state_data.drop(columns='state')
state_data = state_data.to_numpy().squeeze()

window_size = 7

cv_data = np.convolve(state_data, np.ones(window_size), 'same') / window_size

plt.scatter(np.arange(len(state_data)),state_data,s=0.5,label="Original Data")
plt.scatter(np.arange(len(cv_data)),cv_data,s=0.5,label="Convolved Data")
plt.legend()
plt.show()

plt.scatter(np.arange(len(state_data)),state_data,s=0.5,label="Original Data")

pd.DataFrame(cv_data).to_csv(f'../datasets/JHU_CONVOLVED_{state}.csv')


```
